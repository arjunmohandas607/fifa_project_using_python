```python
def convert_wage(wage):
    wage = wage.replace('€', '')
    
    if 'M' in wage:
        return float(wage.replace('M', '')) * 1_000_000
    elif 'K' in wage:
        return float(wage.replace('K', '')) * 1_000
    else:
        return float(wage)

data['Wage'] = data['Wage'].apply(convert_wage)
```

```python
#top5 salary player
top5 = data.nlargest(5, 'Wage')[['Name', 'Wage']]
print(top5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(top5['Name'], top5['Wage'])

plt.xlabel("Player Name")
plt.ylabel("Wage")
plt.title("Top 5 Highest Salary Players")

plt.xticks(rotation=45)
plt.show()
```

```python

def convert_height(x):
    try:
        if isinstance(x, str):
            feet, inches = x.split("'")
            inches = inches.replace('"', '')
            return int(feet) * 30.48 + int(inches) * 2.54
        else:
            return None  
    except:
        return None

data['Height'] = data['Height'].apply(convert_height)


data = data.dropna(subset=['Height'])


top5_tallest = data.nlargest(5, 'Height')[['Name', 'Height']]

print(top5_tallest)
```

```python
#top5 shortest players
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load dataset (change file name if needed)
data = pd.read_csv("fifa_data.csv")

# Step 2: Check first few values (optional)
print(data['Height'].head)
def convert_height(x):
    try:
        if isinstance(x, str) and "'" in x:
            feet, inches = x.split("'")
            inches = inches.replace('"', '')
            return int(feet) * 30.48 + int(inches) * 2.54
        elif isinstance(x, str) and "cm" in x:
            return float(x.replace("cm", ""))
        else:
            return None
    except:
        return None

data['Height'] = data['Height'].apply(convert_height)

data = data.dropna(subset=['Height'])

shortest_players = data.nsmallest(5, 'Height')[['Name', 'Height']]

print("Top 5 Shortest Players:")


```
